# Шаг 4
Сохраните свой прогресс
Когда пользователь нажимает на ✅ кнопка, это означает, что они знают текущее слово на карточке, и это слово следует удалить из списка слов, которые могут появиться.
например, если в french_words.csv было только 3 записи:

chaque,each
parlé,speak
arrivé,come
После того, как пользователь увидел, parlé,speakего следует удалить из списка слов, которые могут снова появиться.

Обновленные данные следует сохранить в новый файл с именем words_to_learn.csv.
например, words_to_learn.csv

chaque, each arrivé, come 3. В следующий раз, когда программа будет запущена, она должна проверить, есть ли файл words_to_learn.csv. Если он существует, программа должна использовать эти слова для нанесения на карточки. Если файл words_to_learn.csv не существует (т. Е. При первом запуске программы), тогда он должен использовать слова вfrench_words.csv

Мы хотим, чтобы наша программа с карточками проверяла нас только на том, чего мы не знаем. Итак, если пользователь нажимает✅ кнопка, это означает, что текущая карта больше не должна появляться.


СОВЕТЫ:
Метод remove () может удалять элементы из списка. например, https://www.w3schools.com/python/ref_list_remove.asp

Вы можете создать новый файл csv из словаря, используя DataFrame.to_csv (), например https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_csv.html

Если вы не хотите создавать индекс для нового CSV, вы можете установить для параметра индекса значение False. например

data.to_csv("filename.csv", index=False)
